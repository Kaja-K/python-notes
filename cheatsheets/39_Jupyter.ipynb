NUMERICAL ANALYSIS AND VISUALIZATION

```python
# Returns descriptive statistics for the 'Unit_Cost' column (mean, std, min, max, etc.)
sales["Unit_Cost"].describe()
```

```python
# Calculates various statistics for the 'Unit_Cost' 
print("Count:  ", sales["Unit_Cost"].count())           # Number of non-null values
print("Median: ", sales["Unit_Cost"].median())          # Middle value
print("Mean:   ", sales["Unit_Cost"].mean())            # Average value
print("Std:    ", sales["Unit_Cost"].std())             # Standard deviation
print("Min:    ", sales["Unit_Cost"].min())             # Minimum value
print("Q1:     ", sales["Unit_Cost"].quantile(0.25))    # 25th percentile
print("Q2:     ", sales["Unit_Cost"].quantile(0.50))    # 50th percentile
print("Q3:     ", sales["Unit_Cost"].quantile(0.75))    # 75th percentile
print("Max:    ", sales["Unit_Cost"].max())             # Maximum value
```

RELATIONSHIP BETWEEN THE COLUMNS - CORRELATION

```python
# Calculate correlation only on numeric columns, ignoring text/date columns
corr = sales.corr(numeric_only=True)
corr
```

```python
# Creates a figure with size 8x8 inches
fig = plt.figure(figsize=(8,8))

# Displays the correlation matrix as a color-coded grid
plt.matshow(corr,cmap="RdBu",fignum=fig.number)
# Sets X-axis labels to column names, rotated vertically
plt.xticks(range(len(corr.columns)),corr.columns,rotation="vertical");
# Sets Y-axis labels to column names
plt.yticks(range(len(corr.columns)),corr.columns);
```

```python
# Creates a scatter plot showing the relationship between 'Quantity_Sold' (X) and 'Sales_Amount' (Y)
sales.plot(kind="scatter",x="Quantity_Sold",y="Sales_Amount", figsize=(6,6))
```

```python
# Creates a scatter plot showing the relationship between 'Unit_Price' (X) and 'Unit_Cost' (Y)
sales.plot(kind="scatter",x="Unit_Price",y="Unit_Cost", figsize=(6,6))
```

Quantity_Sold vs Sales_Amount — points are scattered randomly across the entire chart, meaning there is no correlation between the number of units sold and the total sales amount.
Unit_Price vs Unit_Cost — points form a nearly perfect diagonal line, meaning there is a very strong positive correlation — as the unit price increases, the unit cost increases proportionally.

Conclusion: Unit_Price and Unit_Cost move together closely (likely a fixed margin/markup), while Quantity_Sold has no predictable effect on Sales_Amount.

COLUMN WRANGLING

```python
# Creates a new column by dividing Unit_Cost by Unit_Price to get cost-to-price ratio (0-1)
sales["Cost_Price_Ratio"] = sales["Unit_Cost"] / sales["Unit_Price"]
sales["Cost_Price_Ratio"].head()
```

```python
# Creates a density (KDE) plot showing the distribution of Cost_Price_Ratio
sales["Cost_Price_Ratio"].plot(kind="density",figsize=(14,6))
```

```python
# Creates a histogram showing the frequency distribution of Cost_Price_Ratio
sales["Cost_Price_Ratio"].plot(kind="hist",figsize=(14,6))
```

```python
# Creates a new column by multiplying Unit_Price by Quantity_Sold to get total revenue
sales["Total_Revenue"] = sales["Unit_Price"] * sales["Quantity_Sold"]

# Counts the number of rows where Total_Revenue is not equal to Unit_Cost 
print((sales["Total_Revenue"] != sales["Unit_Cost"]).sum())

# Displays the first 5 values of Total_Revenue
sales["Total_Revenue"].head()
```

```python
# Creates a density (KDE) plot showing the distribution of Total_Revenue
sales["Total_Revenue"].plot(kind="density",figsize=(14,6))
```

```python
# Creates a histogram showing the frequency distribution of Total_Revenue
sales["Total_Revenue"].plot(kind="hist",figsize=(14,6))
```

```python
# Creates a scatter plot showing the relationship between 'Total_Revenue' (X) and 'Sales_Amount' (Y)
sales.plot(kind="scatter", x="Total_Revenue", y="Sales_Amount", figsize=(6, 6))
```

```python
# Displays the first 5 values of Unit_Price before the update
print(sales["Unit_Price"].head())

# Increases all Unit_Price values by 3% (multiplies each value by 1.03)
sales["Unit_Price"] *= 1.03

# Displays the first 5 values of Unit_Price after the update
print(sales["Unit_Price"].head())
```

```python
# Calculates the average Total_Revenue where Region is 'North' AND Sales_Rep is 'David'
sales.loc[(sales["Region"] == "North") & (sales["Sales_Rep"] == "David"), "Total_Revenue"].mean()
```
